```python
# Import libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Load data
car_data = pd.read_csv('Cars93.csv', index_col = 0, keep_default_na=False)
```

### 1. Boxplot

```python
# Boxplot for revs per mile

# Select rows from the dataframe for plotting
#car_data[['Manufacturer', 'Rev.per.mile']].query('Manufacturer == "Suzuki" and Manufacturer =="Audi"')
#car_data

manufacturers = ['Audi', 'Hyundai', 'Suzuki', 'Toyota']
filt = car_data['Manufacturer'].isin(manufacturers)
features_to_plot = car_data.loc[filt, ['Manufacturer' ,'Rev.per.mile']]

sns.boxplot(data = features_to_plot,
            x = features_to_plot['Manufacturer'],
            y = features_to_plot['Rev.per.mile'])
```

**Boxplot question:**  
Which of these manufacturers has the car with the highest revs per mile?


**Boxplot answer:**Toyota


### 2. Histogram

```python
# Histogram of MPG

# Histogram of MPG

# Histogram for MPG in the city
plt.hist(car_data['MPG.city'], bins=10, alpha=0.5, label='MPG in City')

# Histogram for MPG on the highway
plt.hist(car_data['MPG.highway'], bins=10, alpha=0.5, label='MPG on Highway')

# Axes label
plt.xlabel('MPG')
plt.ylabel('Frequency')

# Add legend
plt.legend()

# Show the plot
plt.show()
```

**Histogram question:**  
Is it generally more fuel efficient to drive in the city or on the highway?

**Histogram answer:** The Histogram shows that driving on a highway is more fuel efficient. There are atleast 25 cars that are between the 25-30 miles on a gallon of fuel range. This indicates that a high number of cars achieve this level of fuel efficiency on the highway. Driving in the city shows that there are less than 15 cars in that range. This suggests that fewer cars achieve this level of fuel efficiency in a city.

### 3. Lineplot

```python
# Lineplot (Wheelbase vs turning circle)

sns.lineplot(x='Wheelbase', y='Turn.circle', data=car_data)
```

**Lineplot question:**  
What is this relationship? What happens when the wheelbase gets larger?


**Lineplot answer:**  The relationship is a positive correlation. It means that there is a direct and positive relationship between wheelbase and turn circle. When wheelbase increases, turn circle also increases. This suggests that as the wheelbase of a vehicle becomes larger, it requires a larger turning radius or has a larger turn circle to complete a turn.


### 4. Barplot

```python
# Barplot for mean horsepower

group_by_type = car_data.groupby(by=['Type'])
#group_by_type.head()
average_horse_power = round(group_by_type['Horsepower'].mean(), 0)
average_horse_power.head()
average_horse_power.plot(kind='bar', ylabel='Average horsepower')
```

**Barplot question:**  
Does a larger car mean more horsepower?



**Barplot answer:** Yes. On average, a larger car means more horsepower.
