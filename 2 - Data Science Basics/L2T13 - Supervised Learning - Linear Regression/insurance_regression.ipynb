```python
from sklearn.linear_model import LinearRegression
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# import the data

df = pd.read_csv("insurance.csv")
df.head()
```

```python
# The iloc function allows use to select rows from the dataframe for plotting

x = df.iloc[:, :1].values
y = df.iloc[:, -1].values

plt.scatter(x,y,color = 'b')
plt.xlabel('Age')
plt.ylabel('Insurance Cost')
plt.show()
```

```python
# Fit the model

insurance_model = LinearRegression()
insurance_model.fit(x,y)
```

```python
# Plot the data and model

y_pred = insurance_model.predict(x)
plt.scatter(x,y,color = 'b')
plt.plot(x,y_pred,color = 'r')
plt.xlabel('Age')
plt.ylabel('Insurance Cost')
plt.show()
```

```python
# Predict an unknown value

unknown_x = [[12]]

x_pred = np.append(x, unknown_x).reshape(-1,1)
y_pred = insurance_model.predict(x_pred)

plt.scatter(x,y,color = 'b')
plt.plot(x_pred,y_pred,color = 'r')
plt.xlabel('Age')
plt.ylabel('Insurance Cost')
plt.show()

print("Insurance Cost per age group:", insurance_model.predict(unknown_x))
```
